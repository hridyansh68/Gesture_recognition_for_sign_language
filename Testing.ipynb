```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train  = pd.read_csv("sign_mnist_train/sign_mnist_train.csv")
```

```python
train.head()
```

```python
train.describe()
```

```python
y_train = train['label']
```

```python
X_train = train.drop('label',axis = 1)
```

```python
test = pd.read_csv('sign_mnist_test/sign_mnist_test.csv')
y_test = test['label']
X_test = test.drop('label',axis = 1)
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report
```

```python
from sklearn.tree import DecisionTreeClassifier 
dtree_model = DecisionTreeClassifier().fit(X_train, y_train) 
dtree_predictions = dtree_model.predict(X_test)
```

```python
print(classification_report(y_test,dtree_predictions))
```

```python
from sklearn.svm import SVC 
svm_model_linear = SVC(kernel = 'linear', C = 1).fit(X_train, y_train) 
svm_predictions = svm_model_linear.predict(X_test) 
```

```python
print(classification_report(y_test,svm_predictions))
```

```python
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors = 11).fit(X_train, y_train) 
```

```python
knn_predictions = knn.predict(X_test) 
```

```python
print(classification_report(y_test,knn_predictions))
```

```python
from sklearn.naive_bayes import GaussianNB 
gnb = GaussianNB().fit(X_train, y_train) 
gnb_predictions = gnb.predict(X_test) 
```

```python
print(classification_report(y_test,gnb_predictions))
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier().fit(X_train, y_train)
mlp_predictions = mlp.predict(X_test) 
```

```python
print(classification_report(y_test,mlp_predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier().fit(X_train, y_train)
rfc_predictions = rfc.predict(X_test)
```

```python
print(classification_report(y_test,rfc_predictions))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc = KNeighborsClassifier().fit(X_train, y_train)
knc_predictions = knc.predict(X_test)
```

```python
print(classification_report(y_test,knc_predictions))
```

```python
from sklearn.linear_model import RidgeClassifier
```

```python
rcc = RidgeClassifier().fit(X_train, y_train)
rcc_predictions = rcc.predict(X_test)
```

```python
print(classification_report(y_test,rcc_predictions))
```

```python
from sklearn.linear_model import RidgeClassifierCV
```

```python
rccc = RidgeClassifierCV().fit(X_train, y_train)
rccc_predictions = rccc.predict(X_test)
```

```python
print(classification_report(y_test,rccc_predictions))
```
